**1. Numeric variable: For ./data/airbnb_hw.csv, clean the Price variable as well as you can, and explain the choices you make. How many missing values do you end up with? (Hint: What happens to the formatting when a price goes over 999 dollars, say from 675 to 1,112?)**

```python
import pandas as pd
import numpy as np
import seaborn as sns

df = pd.read_csv('./wrangling/assignment/data/airbnb_hw.csv', low_memory=False)
print(df.shape, '\n')
df.head()
```

```python
price = df['Price']
price.unique()
```

```python
price = df['Price'] # Assigns Price column to df
price = price.str.replace(',', '') # Replaces all the commons with nothing
price = pd.to_numeric(price, errors='coerce') # Converts price to floats and numerics
print(price.unique(), '\n')
print('Total missing: ', sum(price.isnull())) # Sum() counts the number of true that correspond to missing entries
```

Prices over a thousand often include a comma. By removing the comma using the replace function we can ensure that the prices are interpreted as numeric values.

After conversion, there were no missing values after cleaning. Which means that all the price entries were converted into numeric format without any formatting issues.

**2. Categorical variable: For the ./data/sharks.csv data covered in the lecture, clean the "Type" variable as well as you can, and explain the choices you make.**

```python
df = pd.read_csv('./wrangling/data/sharks.csv', low_memory=False)
# df.head()
# df.columns.tolist()

df['Type'].value_counts()
```

```python
type = df['Type']

type = type.replace(['Sea Disaster', 'Boat','Boating','Boatomg'],'Watercraft') # Replace into standardize term of Watercraft
type.value_counts()

type = type.replace(['Invalid', 'Questionable','Unconfirmed','Unverified','Under investigation'],np.nan) # Mark these as missing
type.value_counts()

df['Type'] = type # Clean version
del type

df['Type'].value_counts()
```

Replacing terms like 'Sea Disaster', 'Boat', 'Boating', and 'Boatomg' with a standardized term like 'Watercraft' helps to reduce variability in the data.

```python
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].replace(['UNKNOWN', 'F','M','2017'],np.nan) # Uncleaned data values
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].replace('y','Y') # Uncleaned data values
pd.crosstab(df['Type'],df['Fatal (Y/N)'],normalize='index')
```

```python
length = df['ImposedSentenceAllChargeInContactEvent']
type = df['SentenceTypeAllChargesAtConvictionInContactEvent']

# print( length.unique()  , '\n')
length = pd.to_numeric(length,errors='coerce')
length_NA = length.isnull()
print( np.sum(length_NA),'\n')

print( pd.crosstab(length_NA, type), '\n')

length = length.mask( type == 4, 0)
length = length.mask( type == 9, np.nan)

length_NA = length.isnull()
print( pd.crosstab(length_NA, type), '\n')
print( np.sum(length_NA),'\n')

df['ImposedSentenceAllChargeInContactEvent'] = length
del length, type
```
